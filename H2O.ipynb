```python
import h2o
from h2o.automl import H2OAutoML
h2o.init(max_mem_size = 8, port=54321)
import pandas as pd
import warnings
from IPython.display import display

warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', 500)

col_to_drop = ['Valor dólar informal semestral', 'identificador', 'Año', 'Semestre', 'target']

train_data = pd.read_csv('train_data.csv', low_memory=False)
train_data['target'] = train_data['Valor dólar informal semestral']*train_data['Salario mensual (en tu moneda local)']
test_data = pd.read_csv('features.csv', low_memory=False)


for_prediction_h2o = h2o.H2OFrame(test_data.drop(columns=['Valor dólar informal semestral', 'identificador', 'Año', 'Semestre']))
df = h2o.H2OFrame(train_data.drop(columns=['Salario mensual (en tu moneda local)', 'Valor dólar informal semestral', 'identificador', 'Año', 'Semestre']))


aml = H2OAutoML(max_runtime_secs = 300, project_name = "RGA", stopping_metric='MSE')
aml.train(y = 'target', training_frame = df, leaderboard_frame = df)
pred = aml.predict(for_prediction_h2o)

data_as_df = h2o.as_list(pred).rename(columns={'predict': 'target'})
test_data = test_data.join(data_as_df)
test_data[['identificador', 'target']].to_csv('rga_h2o.csv', index=False)
```
